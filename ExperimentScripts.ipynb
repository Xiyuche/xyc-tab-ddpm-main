## Mask
- generate mask
- get known dataset
- get unknown dataset

```python
# Load your dataset
X_num_train = np.load('ExperimentLocalData/X_num_train.npy')  # replace with your actual file path

# Create a random mask where 1 represents known data and 0 represents unknown data
mask = np.random.randint(2, size=X_num_train.shape)

# display mask
mask
```

```python
# Known dataset: elementwise multiplication of the dataset with the mask
known_dataset = X_num_train * mask

# Unknown dataset: elementwise multiplication of the dataset with the inverse of the mask
unknown_dataset = X_num_train * (1 - mask)
```
